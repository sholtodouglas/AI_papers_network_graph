```python
import sys
import pickle
import pandas as pd
import time
import collections
from tqdm import tqdm
SPARSITY = 2000
```

```python
# df_msft = pickle.load((open('df_msft_all.pkl', 'rb')))
# df_msft
df_full = pickle.load((open('df_full.pkl', 'rb')))
df_full
```

```python

def count_refs(df):
    count_list = [] # total count of each ref
    for row in tqdm(df.itertuples()):
        
        paper = row.msft_id
        refs = row.msft_refs

        for r in refs:
            
            count_list.append(r)
         
        
    counter = collections.Counter(count_list)
    return counter

counter = count_refs(df_full)
```

```python
counter.most_common()
```

```python
def get_top_n_papers(n, counter):
    top_n = []
    for c in counter.most_common(n):
            top_n.append(c[0]) # put the code
    return top_n

popular_papers = get_top_n_papers(SPARSITY, counter) # top n papers

print(len(popular_papers))
print(len(counter.most_common()))
```

```python
popular_papers.remove('N')
popular_papers.remove('A')
popular_papers
```

```python
title_dictionary = dict(zip(list(df_full['msft_id']), list(df_full['title'])))
year_dictionary = dict(zip(list(df_full['msft_id']), list(df_full['year'])))
```

```python
source = []
target = []
def create_links(df):
    for row in tqdm(df.itertuples()):
        paper = row.msft_id
        refs = row.msft_refs

        for r in refs:
            if r in popular_papers and paper in popular_papers:
                source.append(paper)
                target.append(r)
                
create_links(df_full)
       
```

```python
graph_codes = []
graph_titles = []
graph_years = []

def create_nodes(df):
    for row in tqdm(df.itertuples()):
        
        
        list_of_ids = [row.msft_id]  # create a list of all the codes

        if row.msft_refs != 'NA':
            list_of_ids = list_of_ids + row.msft_refs
            
            
        for code in list_of_ids:
            # fill out the dictionary
            if code not in graph_codes and code in popular_papers and (code in source or code in target): #i.e, we haven't already made a node for it

                if code in title_dictionary.keys():
                    title = title_dictionary[code]
                else:
                    title = code ############ for testing, assume some title.
                    
                if code in year_dictionary.keys():
                    year = year_dictionary[code]
                else:
                    year = 2018 ############ for testing, assume some year.
                
                
                ############
                graph_codes.append(code)
                graph_titles.append(title)
                graph_years.append(year)
                
create_nodes(df_full)
print(len(graph_codes))
   
```

```python
import math
degree = []
for uid in graph_codes:
    # let degree be num incoming edges since we only care about it being referenced
    # as a metric for popularity and not how many times it references others
    #degree.append(target.count(uid) + 1) 
    degree.append(2*(target.count(uid)) + 0.1) 
len(degree)                                            
```

```python
JSON = "{" + '"nodes":' + nodes.to_json(orient = 'records') + ","
JSON = JSON + '"links":' + links.to_json(orient = 'records') + '}'
with open('g2000.json', 'w') as f:
    f.write(JSON)
```
