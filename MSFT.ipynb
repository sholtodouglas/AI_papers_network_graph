```python
def msft_format(title):
    # make lower case and replace all alphanumeric with a space for microsoft formatting
    msft_title = "".join([ c if c.isalnum() else " " for c in title.lower() ])
    #replace any multi spaces with a single space for msft format
    msft_title = ' '.join(msft_title.split())
    return msft_title
```

```python
## Now, lets add all the highly referenced papers and their references. 



def count_refs(df):
    id_2_title = dict(zip(df['msft_id'], df['title']))
    count_list = []
    for row in tqdm(df.itertuples()):
        paper = row.msft_id
        refs = row.msft_refs

        for r in refs:
            count_list.append(r)
            
            
    counter =collections.Counter(count_list)
    
    frequency  = counter.most_common()
    
    freq_list = []
    for tup in frequency:
        tup = list(tup)
        
        if tup[0] in id_2_title:
            tup[0] = id_2_title[tup[0]]
            
        freq_list.append(tup)
        
    return counter, freq_list

counter, frequency = count_refs(df_msft)


```

```python
# so technically we are doing one recursion now. #from the most to least common papers. Getting titles and refs. 

titles = []
msft_refs = []
msft_ids= []
years = []
```

```python


for f in frequency:
    # we don't already have it through arxiv and getting it's title there. 
    if not isinstance(f[0], str) and f[0] not in msft_ids:
        title, refs, year = get_microsoft_title_and_refs_by_ID(f[0])
        
        titles.append(title)
        msft_refs.append(refs)
        years.append(year)
        msft_ids.append(f[0])
        print(f[0],title)
        time.sleep(0.5)
```

```python
seen = dict()
sub_code_dict = dict()
rows_to_drop = []
for ix, row in df_full.iterrows():
    
    # if we've seen it before
    if row.title in seen:
        #get the info of the previous entry
        dup_id, dup_refs, dup_ix = seen[row.title]
        if row.msft_refs != 'NA':
        #update previous entry with our references
            if dup_refs == 'NA':
                df_full.msft_refs.iloc[dup_ix] = row.msft_refs
            else:
                df_full.msft_refs.iloc[dup_ix] = row.msft_refs + dup_refs
            
            #remove the duplicate row
            #df_full.drop(df_full.index[ix], inplace=True)
            rows_to_drop.append(ix)
            #prepare to replace all references to the duplicate 
            sub_code_dict[row.msft_id] = dup_id
            print(row.title)
        
    else:
        seen[row.title] = (row.msft_id, row.msft_refs, ix)
            
```

```python
df_full.drop(df_full.index[rows_to_drop], inplace=True)
df_full = df_full.reindex()
```

```python
new_refs_col = []
for ix, row in tqdm(df_full.iterrows()):
    refs = row.msft_refs
    for idx, item in enumerate(refs):
        if item in sub_code_dict:
            refs[idx] = sub_code_dict[item]
            
    new_refs_col.append(refs)
    
df_full['msft_refs'] = new_refs_col
    
```
